```python
df = pd.read_csv('/content/CORRUPTION INDEX 2022.csv')

# Membuat bar chart
plt.figure(figsize=(40, 12))
colors = ['#FAEBD7' for _ in range(len(df))]
colors[-69] = '#D2691E'
plt.bar(df['Country '], df['Score'], color=colors)  # Memperbaiki nama kolom jika perlu
plt.title('Corruption Perceptions Index')
plt.xlabel('Nama Negara')
plt.ylabel('Score Indeks Korupsi')
plt.xticks(rotation=90, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

Berdasarkan dari visualisasi data diatas dapat dilihat bahwa terdapat Corruption Perceptions Index untuk sumbu Y, sedangkan nama-nama negara di dunia untuk sumbu X. Pada Corruption Perceptions Index menggunakan skala 0(Korupsi yang tinggi) hingga 100(Korupsi yang rendah). Terlihat pada gambar bahwa negara Denmark menjadi negara dengan rank ke-1 dengan score Corruption Perceptions Index adalah 90, sedangkan untuk negara Somalia menjadi negara dengan rank ke-180 dengan score Corruption Perceptions Index adalah 12. Hal ini menunjukkan bahwa Denmark merupakan negara yang korupsinya sangat rendah dan Somalia menjadi negara yang korupsinya yang sangat tinggi. Adapun Indonesia berada pada rank ke-110 dengna score Corruption Perceptions Index adalah 34.
