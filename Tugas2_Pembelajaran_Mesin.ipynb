```python
df = pd.read_csv('/content/CORRUPTION INDEX 2022.csv')

# Membuat bar chart
plt.figure(figsize=(40, 12))
colors = ['#FAEBD7' for _ in range(len(df))]
colors[-69] = '#D2691E'
plt.bar(df['Country '], df['Score'], color=colors)  # Memperbaiki nama kolom jika perlu
plt.title('Corruption Perceptions Index')
plt.xlabel('Nama-Nama Negara di Dunia')
plt.ylabel('Score Corruption Perceptions Index')
plt.xticks(rotation=90, ha='right', fontsize=10)
plt.yticks(range(0,101,20))
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi data di atas merupakan visualisasi dari Corruption Perceptions Index, dapat dilihat bahwa sumbu Y menunjukkan Score Corruption Perceptions Index, sedangkan sumbu X menunjukkan Nama-Nama Negara di Dunia. Skala Corruption Perceptions Index (CPI) berkisar dari 0 (korupsi tinggi) hingga 100 (korupsi rendah).

Pada gambar, terlihat bahwa Denmark menempati peringkat pertama dengan skor CPI 90, sedangkan Somalia berada di peringkat 180 dengan skor CPI 12. Hal ini menunjukkan bahwa Denmark merupakan negara dengan tingkat korupsi yang sangat rendah, sedangkan Somalia memiliki tingkat korupsi yang sangat tinggi. Indonesia berada di peringkat 110 dengan skor CPI 34.

Meskipun Denmark menempati peringkat pertama dalam CPI tahun 2022, masih terdapat ruang untuk perbaikan. Diperlukan upaya berkelanjutan dari semua pihak untuk memerangi korupsi dan menciptakan masyarakat yang lebih adil dan sejahtera.
